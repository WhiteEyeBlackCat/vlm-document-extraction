```python
import os
import random
import csv
```

```python
root = "./"

folder_dict = {
    "0314": root + "0314",
    "Sunrisetek-1": root + "Sunrisetek-1",
    "Sunrisetek-2": root + "Sunrisetek-2",
}

results = []

for folder_name, folder_path in folder_dict.items():
    pdf_files = sorted([f for f in os.listdir(folder_path) if f.lower().endswith(".pdf")])

    rng = random.Random(42)
    sampled = rng.sample(pdf_files, 15)

    for i, f in enumerate(sampled, 1):
        results.append([folder_name, i, f])

with open("sampled_pdfs_seed42.csv", "w", newline="", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(["folder", "sample_id", "filename"])
    writer.writerows(results)

print("已存成 sampled_pdfs_seed42.csv")
```
