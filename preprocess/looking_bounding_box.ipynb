```python
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path
```

```python
%matplotlib widget

img_path = "/home/wenbohao/ITRI_I-A/data/jpg/Sunrisetek-2/4115095649/page-1.jpg"
img = Image.open(img_path)

points = []

def onclick(event):
    if event.xdata is not None and event.ydata is not None:
        x, y = int(event.xdata), int(event.ydata)
        points.append((x, y))
        print(f"({x}, {y})")

fig, ax = plt.subplots(figsize=(8, 10))
ax.imshow(img)
ax.set_xlim(0, img.size[0])
ax.set_ylim(img.size[1], 0)
ax.grid()
fig.canvas.mpl_connect("button_press_event", onclick)
plt.show()
```

```python
img_path = Path("/home/wenbohao/ITRI_I-A/data/jpg/Sunrisetek-2/4115095650/page-1.jpg")
img = Image.open(img_path)

boxes = {
    "head": (51, 130, 2315, 450),
    "ship": (41, 465, 1126, 902),
    "general_info": (1226, 460, 2279, 925),
    "lower_half": (41, 975, 2247, 3363),
}

for name, box in boxes.items():
    cropped = img.crop(box)
    plt.figure(figsize=(8, 4))
    plt.imshow(cropped)
    plt.title(name)
    plt.axis("off")
    plt.show()
    plt.close()
```

```python
base_path = Path("/home/wenbohao/ITRI_I-A/data/jpg")


for fold in ["0314", "Sunrisetek-1", "Sunrisetek-2"]:
    fold_path = Path(f"/home/wenbohao/ITRI_I-A/data/jpg/{fold}")
    for img_path in sorted(fold_path.rglob("*.jpg")):
        if any(tag in img_path.stem for tag in boxes):
            continue

        img = Image.open(img_path)

        for name, box in boxes.items():
            out_path = img_path.parent / f"{img_path.stem}_{name}.jpg"
            cropped = img.crop(box)
            cropped.save(out_path, quality=95)
        
        print(f"Done: {img_path}")
```
